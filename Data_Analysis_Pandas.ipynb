```python
#A function that opens the file path for the dataset
def open_file(fpath):
  data = pd.read_csv(fpath)
  return data
```

```python
#Opens the file from the given path and turns it into a dataframe
fpath = f'{folder}/ncaa-team-data.csv'
data = open_file(fpath)
data
```

```python
#A function turning arguments into a numeric type (for division)
def numeric(data):
  data['w'] = pd.to_numeric(data['w'])
  data['l'] = pd.to_numeric(data['l'])
  return data
```

```python
#Calculating the winrate for every team
def win_rate(data):
  wins = data['w'] #assigning the 'w' column to a str
  losses = data['l'] #assinging the 'l' column to a str
  total = wins + losses #adding up wins and losses
  if total > 0: #To parse through and find teams that have a winrate of '0' (prevents the error of "float divison by zero")
    data_div = wins/total
    return data_div
```

```python
#A function that applies a new column 'winrate' to the dataset
def apply_win_rate(data):
  data['winrate'] = data.apply(win_rate, axis=1) #defining the new column "winrate" and applying it through the .apply() function, axis=1 represents columns - in this case, "winrate" is being applied as a column
  return data
```

```python
#Testing the function to see if the column is applied to the dataset
data_winrate = apply_win_rate(data)
data_winrate
```

```python
#Grouping by multiple columns. In this case, I am keeping the original columns of coaches and year, and finding the mean for the winrate (to determine the average winrate for each coach and which year).
new_data = data.groupby(['school','coaches', 'season']).agg({'winrate': ['mean']}) 
#Creating a new column for the mean winrate
new_data.columns = ['winrate'] 
#Generating a new dataframe or series with the index reset and dropping "NaN" values in the dataset
new_data = new_data.reset_index().dropna()
```

```python
#Sorts by the highest average winrate and grabs the first 5 (in this case it would be the top five)
top_5 = new_data.sort_values(by='winrate', ascending=False).head(5)
#Sorts by the least average winrate and grabs the first 5 (in this case it would be the worst five)
last_5 = new_data.sort_values(by='winrate', ascending=True).head(5)
```

```python
#Creates a new dataset with only the top 5 best winrates and the worst 5 winrates
top_last_5 = pd.concat([top_5, last_5], axis=0)
top_last_5
```
